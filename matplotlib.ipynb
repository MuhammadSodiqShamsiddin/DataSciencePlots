```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sonlar = np.arange(10)
```

```python
plt.plot(sonlar)
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize = (10, 4))
plt.plot(x, y, linewidth = 2, linestyle = '--', color = 'g', marker = '8')
plt.title("sin(x) Function")
plt.xlabel("Radian Arguments")
plt.ylabel("Functions values")
plt.grid(color = 'k', linestyle = '--')
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize = (16, 6))

# sin(x) grafigi
plt.plot(x, sin,
        linewidth = 1.5,
        linestyle = '-',
        color = "#8E44AD",
        alpha = 0.7,
        marker = 'h')

# cos(x) grafigi
plt.plot(x, cos,
        linewidth = 1.5,
        linestyle = '-',
        color = "c",
        alpha = 0.7,
        marker = 'v')
plt.title('sin(x) and cos(x) Graphics')
plt.xlabel("Radian Arguments")
plt.ylabel("Function Values")
plt.legend(['sin(x) function', 'cos(x) function'], loc = 1, fontsize = 'x-large') #PLOT info
plt.grid(color = 'tomato', alpha = 0.5, linestyle = '--') # Setkani o'zgartirish
plt.savefig("graphic1.png")
plt.show()
```
