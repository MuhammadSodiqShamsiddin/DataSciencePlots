```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year>2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(data = df, x = 'price')
plt.show()
```

```python
sns.displot(data = df, hue = 'transmission', x = 'price', height = 5, aspect = 2, multiple = 'stack')
plt.show()
```

```python
sns.displot(data = df, col = 'transmission', x = 'price', height = 5, aspect = 2)
plt.show()
```

```python
sns.displot(data = df, kind = 'kde', height = 5, aspect = 2, x = 'price')
plt.show()
```

```python
sns.displot(data = df, kde = True, height = 5, aspect = 2, x = 'price')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.kdeplot(data = df, x = 'price', hue = 'fuelType', shade = True)
plt.show()
```

```python
sns.displot(data = df, kind = 'kde', height = 5, aspect = 2, x = 'price', hue='fuelType')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(data = df, x = 'price', kind = 'ecdf', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.grid(which = 'both') #minor major both
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data = df, x = 'price', hue = 'transmission', kind = 'kde', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(data = df, x = 'price', col = 'fuelType', kind = 'kde', height = 5, aspect = 2)
plt.show()
```
