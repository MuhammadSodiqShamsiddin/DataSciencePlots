```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col = 0)
df
```

```python
plt.figure(figsize = (16, 6))
sns.barplot(x = df.index, y = df['Aholisi'])
plt.xticks(rotation = 90)
plt.title("O'zbekiston Respublikas tarkibidagi hududlar aholi soni")
plt.xlabel("Hudud")
plt.ylabel("Aholi soni (10^6)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (16, 6))
sns.barplot(x = df.index, y = df['Maydoni (kv.km)'])
plt.xticks(rotation = 90)
plt.title("O'zbekiston Respublikas tarkibidagi hududlar maydoni")
plt.xlabel("Hudud")
plt.ylabel("Maydoni (kv.km)")
plt.grid()
plt.show()
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col = 0, parse_dates = [0], usecols = [0, 1])
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).max()
df2
```

```python
df2.unstack(level = 1)
```

```python
plt.figure(figsize = (12, 6))
sns.heatmap(data = df2.unstack(level=0), cmap = "Blues", linewidths = 1, square = True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels = xticks_labels)
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x=df['Month'], y = df['AA'])
plt.title("American Airlines kompaniyasi samolyotlarining kechikish vaqti")
plt.ylabel("Minutes")
plt.show()
```

```python
df
```

```python
df2 = df[['Month', 'AA']]
df2['values'] = df2['AA']
df2['Variable'] = 'AA'
df2
```

```python
sns.catplot(data = df2, kind = 'bar', x = 'Month', y = 'values', hue = 'Variable')
plt.show()
```

```python
df2 = pd.melt(df, id_vars=['Month'], value_vars=['AA', 'AS', 'DL', 'WN', 'MQ'])
df2.head()
```

```python
sns.catplot(data = df2, kind = 'bar', x = "Month", y = 'value', hue = "variable")
plt.show()
```
