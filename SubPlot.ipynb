```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

fig1 = plt.figure(figsize = (16, 6))

plt.subplot(2, 1, 1)
# sin(x) grafigi
plt.plot(x, sin,
        linewidth = 1.5,
        linestyle = '-',
        color = "#8E44AD",
        alpha = 0.7,
        marker = 'h')
plt.subplot(2, 1, 2)
# cos(x) grafigi
plt.plot(x, cos,
        linewidth = 1.5,
        linestyle = '-',
        color = "c",
        alpha = 0.7,
        marker = 'v')
plt.title('sin(x) and cos(x) Graphics')
plt.xlabel("Radian Arguments")
plt.ylabel("Function Values")
plt.legend(['sin(x) function', 'cos(x) function'], loc = 1, fontsize = 'x-large') #PLOT info
plt.grid(color = 'tomato', alpha = 0.5, linestyle = '--') # Setkani o'zgartirish
plt.savefig("graphic2.png")
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df = df[df.year>2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 10))
sns.histplot(ax = ax[0], data = df, x = 'price')
sns.scatterplot(ax = ax[1],data = df, x = 'price', y = 'year')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (16, 12))
fig.suptitle("Graphic3")
sns.histplot(ax = ax[0, 0], data = df, x = 'price')
sns.histplot(ax = ax[0, 1], data = df, x = 'mileage')
sns.scatterplot(ax = ax[1, 0], data = df, x = 'year', y = 'price')
sns.scatterplot(ax = ax[1, 1], data = df, x = 'mileage', y = 'price')
ax[0, 0].set_title("Price Distribution")
ax[0, 1].set_title("Mileage Distribution")
ax[1, 0].set_title("Year vs Price Correlation")
ax[1, 1].set_title("Mileage vs Price Correlation")
plt.savefig("graphic3.png")
plt.show()
```
